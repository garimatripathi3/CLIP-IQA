```python
import cv2
import matplotlib.pyplot as plt
import skimage
from skimage import io, color

import PIL
from PIL import Image
import os
import numpy as np
```

```python
#Creating No librarary directory for the images, which are converted to grayscale without the help of any inbuilt
#library.
directory_path = "../images/No library"
if not os.path.exists(directory_path):
    os.makedirs(directory_path)
```

```python
#Subplot of all four images using matplotlib and opencv library.
for i in range(1,5):
   im_path = f"../images/sam{i}.jpg"
   image = cv2.imread(im_path)

   
   if image is not None:
        plt.subplot(2,2,i)
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        

```

```python
#Converting to grayscale using sum formula  (r+g+b)/3
for i in range(1,5):
    im_path = f"../images/sam{i}.jpg"
    image = cv2.imread(im_path)
    gray_scale = np.sum(image, axis = 2) //3
    output_path = f"../images/No library/sam{i}.jpg"
    cv2.imwrite(output_path, gray_scale)
```

```python
#Creating Scikit-image directory for the images, which are converted to grayscale with the help of Scikit- image
#library.
directory_path = "../images/Scikit-image"
if not os.path.exists(directory_path):
    os.makedirs(directory_path)
```

```python
#Converting to grayscale using Scikit-image library
for i in range(1,5):
    im_path = f"../images/sam{i}.jpg"
    image = io.imread(im_path)
    gray_image = color.rgb2gray(image)
    output_path = f"../images/Scikit-image/sam{i}.jpg"
    io.imsave(output_path, gray_image)
```

```python
#Creating PIL directory for the images, which are converted to grayscale with the help of PIL library.
directory_path = "../images/PIL"
if not os.path.exists(directory_path):
    os.makedirs(directory_path)
```

```python
#Converting to grayscale using PIL library
for i in range(1,5):
    im_path = f"../images/sam{i}.jpg"
    image = Image.open(im_path)
    gray_image = image.convert("L")
    output_path = f"../images/PIL/sam{i}.jpg"
    gray_image.save(output_path)
```

```python
#Code for taking webcam image
web_image = cv2.VideoCapture(0)
ret, frame = web_image.read() 
#frame contain image data
# ret is boolean variable, denoting whether image capture was successful or not
```

```python
#Creating webcam directory to store webcam image
directory_path = "../images/Webcam"
if not os.path.exists(directory_path):
    os.makedirs(directory_path)
```

```python
#Releasing the camera from computer resource
web_image.release()

if not ret :
    print("Unable to capture webcam image")
else:
    #Saving image to Webcam directory.
    cv2.imwrite(f"../images/Webcam/webcam.jpg",frame)
    print("Webcam image captured  successfully!")
    
```
